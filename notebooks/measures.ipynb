```python
import os

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.metrics import classification_report

from reconocimiento_crotales.PretrainedReader import PretrainedReader
```

```python
plt.rcParams['figure.facecolor'] = 'white'
```

```python
df = pd.read_csv('images/GroundTruth.csv', sep=';')
```

```python
df['file'] = df['NUM_DOCUMENTO'].apply(
    lambda x: os.path.join(
        'images', 'TestSamples', f"{x.replace('DOCUMENTO_0', '')}.TIF")
)
```

```python
df['Real'] = df['Real'].shift(1)
```

```python
reader = PretrainedReader()
```

```python
def predict(f):
    try:
        r = reader.process_image(f).get_value()
        print(r)
        return str(r)
    except:
        return None
    
df['predicted'] = df['file'].apply(lambda f: predict(f))
```

```python
df.sample(10)
```

```python
def str2int(x):
    if x:
        return int(x)
    return None

correct_pred = sum(df['Real'] == df['predicted'].apply(str2int))
correct_pred
```

```python
len(df['Real'])
```

```python
fig1, ax1 = plt.subplots()
ax1.pie(
    [correct_pred, len(df) - correct_pred],
    explode=[.1, 0],
    labels=['Correctamente identificados.', 'Incorrectamente identificados.'], 
    autopct='%1.1f%%',
    shadow=True, startangle=90
)
ax1.axis('equal')
```

```python
from collections import defaultdict

aux = pd.DataFrame(columns=['real', 'predicted'])
for _, row in df.iterrows():
    if not row['predicted']:
        continue
    r = str(int(row['Real']))
    p = str(int(row['predicted']))
    for i in range(1, len(p) + 1):
        try:
            aux = aux.append({'real': r[-i], 'predicted': p[-i]}, ignore_index=True)
        except IndexError:
            continue
```

```python
print(classification_report(aux['real'], aux['predicted']))
```

```python
def plot_count(c, **kwargs):
    foo = pd.Series(np.zeros(10), dtype=np.int64, index=[str(i) for i in range(10)])
    foo.update(pd.value_counts(c))
    sns.barplot(foo.index, foo.values)

g = sns.FacetGrid(aux, col="real", col_wrap=5, height=2)
g.map(plot_count, 'predicted');
```
